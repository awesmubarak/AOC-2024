```python
with open('5-1.txt') as f:
    p_input = f.read()
```

```python
raw_rules, raw_updates = p_input.split("\n\n")
# p for prequisite, d for dependant
rules = [{'p':int(x), 'd':int(y)} for line in raw_rules.split("\n") for x, y in [line.split("|")]]
updates = [[int(x) for x in line.split(",")] for line in raw_updates.split("\n")]
```

```python
invalid_updates = []
for update in updates:
    for rule in rules:
        preq = rule['p']
        dep = rule['d']
        if preq in update and dep in update:
            if update.index(preq) >= update.index(dep):
                invalid_updates.append(update)
                break
print(invalid_updates)
```

```python
fixed_updates = []
for update in invalid_updates:
    update_copy = update.copy()
    needs_fixing = True
    
    while needs_fixing:
        needs_fixing = False
        for rule in rules:
            preq = rule['p']
            dep = rule['d']
            if preq in update_copy and dep in update_copy:
                index_preq = update_copy.index(preq)
                index_dep = update_copy.index(dep)
                if index_preq >= index_dep:
                    update_copy[index_preq], update_copy[index_dep] = update_copy[index_dep], update_copy[index_preq]
                    needs_fixing = True
                    break
    
    fixed_updates.append(update_copy)
```

```python
print(sum([int(li[len(li)//2]) for li in fixed_updates]))
```
